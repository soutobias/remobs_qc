```python
conn = connect_database_remo('PRI')
buoy_id = bmo_on(conn)
id = buoy_id['buoy_id'][0]
last_date_raw = check_last_date(conn, 'bmo_br_current', id)
last_date_raw = last_date_raw[0][0]
time_interval = 24 # hours interval of data to be qualified
bmo_general = get_data_table_db(conn, id, last_date_raw, 'bmo_br_current','ALL')
bmo_general.set_index('date_time', inplace = True)
bmo_general.head()
```

```python
x = list(bmo_general.columns[4:-1])
x[0][0:4] == 'cspd'
```

```python
df = bmo_general.copy()
def definition_flag_pandas_adcp(df):

    flag_data = df [list(df.columns[4:-1])] * 0

    flag_data = flag_data.replace(np.nan, 0).astype(int)

    return flag_data

flag_data = definition_flag_pandas_adcp(df)
# flag_data.head()
parameters = list(df.columns[4:-1])
```

```python
df
```

```python
for parameter in parameters:
    # missing value checks
    if parameter[0:4] == 'cspd':
        df.loc[df[parameter] != -9999, parameter] = df[parameter] * 1000
    flag_data = qc.mis_value_check(df.copy(), limits.mis_value_limits, flag_data.copy(), parameter)
    # coarse range check
    flag_data = qc.range_check(df.copy(), limits.range_limits, flag_data.copy(), parameter)
    # soft range check
    flag_data = qc.range_check_climate(df.copy(), limits.climate_limits, flag_data.copy(), parameter)
```

```python
flag_data = qc.tcontinuityadcpcheck(df.copy(), limits.adcp_limits, 3, flag_data.copy(), ["cspd1", "cdir1"])
```
